```python
import pandas as pd
```

```python
scores = pd.read_csv('../data/lifelines_per_zone_20200218.csv')
```

```python
scores
```

```python
scores['score_n'] = scores['lifeline_1']/scores['lifeline_1'].sum() + \
                  scores['lifeline_2']/scores['lifeline_2'].sum() + \
                  scores['lifeline_3']/scores['lifeline_3'].sum() + \
                  scores['lifeline_4']/scores['lifeline_4'].sum() + \
                  scores['lifeline_5']/scores['lifeline_5'].sum() + \
                  scores['lifeline_6']/scores['lifeline_6'].sum()
```

```python
scores['score_n_weighted'] = .2 * scores['lifeline_1']/scores['lifeline_1'].sum() + \
                  .2 * scores['lifeline_2']/scores['lifeline_2'].sum() + \
                  .3 * scores['lifeline_3']/scores['lifeline_3'].sum() + \
                  .1 * scores['lifeline_4']/scores['lifeline_4'].sum() + \
                  .1 * scores['lifeline_5']/scores['lifeline_5'].sum() + \
                  .1 * scores['lifeline_6']/scores['lifeline_6'].sum()
```

```python
scores['score_weighted'] = .2 * scores['lifeline_1'] + \
                  .2 * scores['lifeline_2'] + \
                  .3 * scores['lifeline_3'] + \
                  .1 * scores['lifeline_4'] + \
                  .1 * scores['lifeline_5'] + \
                  .1 * scores['lifeline_6']
```
